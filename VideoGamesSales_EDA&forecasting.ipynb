<h2><b>ARIMA Forecasting<b><h2>

```python
series=df_year_avg
series_portion=series["2000":"2015"]
```

```python
#Check For Stationarity
from statsmodels.tsa.stattools import acf, pacf, adfuller, kpss

def test_stationarity(series): 
    #adf test
    result=adfuller(series)
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    print("Critical Values:")
    for key, value in result[4].items():
        print(f"   {key}: {value}")

    #kpss test
    result2=kpss(series)
    print(f"KPSS Statistic: {result2[0]}")
    print(f"p-value: {result2[1]}")
    print("Critical Values:")
    for key, value in result2[3].items():
        print(f"   {key}: {value}")

test_stationarity(series_portion)
```

as we have seen from the graphs and from the adf test, the series is not stationary, so we will have to apply first differncing on it to make it stations and be able to form our ARIMA model on it.

```python
diff_series=series_portion.diff().dropna()
plt.plot(diff_series)
test_stationarity(diff_series)
```

still not stationary so we apply one more differncing

```python
diff_series2=diff_series.diff().dropna()
plt.plot(diff_series2)
test_stationarity(diff_series2)
```

we apply differncing again.

```python
diff_series3=diff_series2.diff().dropna()
plt.plot(diff_series3)
test_stationarity(diff_series3)
```

after 3 consecutive differncings, we can see graphically and also through the tests that thet outcome series is stationary (constant mean and variance and covariance independent of t).

```python
n_lags=12
fig,axs=plt.subplots(1,2,figsize=(15,6))

axs[0].grid(visible=True,axis="y",zorder=0)
axs[0].bar(np.arange(0,n_lags),acf(diff_series3)[:n_lags],zorder=3)
axs[0].set_title("ACF of 3rd differenced series")
axs[0].set_xlabel("lags K")
axs[0].set_ylabel("ACF Values")

axs[1].grid(visible=True,axis="y",zorder=0)
axs[1].bar(np.arange(0,6),pacf(diff_series3)[0:6],zorder=3)
axs[1].set_title("PACF of 3rd differenced series")
axs[1].set_xlabel("lags K")
axs[1].set_ylabel("PACF Values")
```

For a significance level=0.2, looking at the ACF plot we see a high negative correlation at k=1, bu at also at larger k such as k=8, 9 and 10 but we will treat them as random fluctuations since other K between 2 and 8 are very low, plus it's not likely to be a seasonality case here. And for the PACF we see high patial autocorrelation at k=1 and k=2.
Hence the model we will start with is gonna be ARIMA(1,3,2)

```python
# fit ARIMA model
from statsmodels.tsa.arima.model import ARIMA

fig = plt.figure(figsize=(11,6))
model = ARIMA(series_portion, order=(1,3,2)) 
ax = plt.gca()
results = model.fit() 
plt.plot(series_portion)
plt.plot(results.fittedvalues, color='red')
ax.legend(['Global Sales', 'Forecast'])

print(results.summary())
```

```python
# Actual values up to forecast point
plt.plot(series.loc[:"2015-01-01"], label='Actual')

# Forecasted values
forecast = results.predict(start="2015-01-01", end="2017-01-01", dynamic=False)
plt.plot(forecast, color='red', label='Forecast')

# Enhancements
plt.title('ARIMA Forecast of Car Counts')
plt.xlabel('Time')
plt.ylabel('Car Count')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

We mannaged to forecast with a lead time l=1, it looks likes global sales are expected to decrease given past years have shown a downward trend, also it is import to note that the order paramters we chose for the arima model might not be the most optimal or best selection to adapt here. we can automate this using auto_arima function that tries to find the best possible set of order that minimizes AIC an increases the accuracy.
